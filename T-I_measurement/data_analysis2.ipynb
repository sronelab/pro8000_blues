```python
import json
import matplotlib.pyplot as plt
import numpy as np

# load data and see if it makes sense
fname = 'data/20250826_103358.json'
with open(fname) as f:
    data = json.load(f)

fig, axes = plt.subplots(3, 1)
for i, laser in enumerate(data):
    current = laser['I_measured']
    power = laser['P_measured']
    axes[i].plot(current, power, color='k')
    axes[i].set_title(laser['name'])
    axes[i].set_xlabel('Current (mA)')
    axes[i].set_ylabel('Power (a.u.)')
fig.tight_layout()
```

```python
# Plot first derivative of the curve

fig, axes = plt.subplots(3, 1)
for i, laser in enumerate(data):
    current = laser['I_measured']
    power = laser['P_measured']

    deriv = np.diff(power, n=1)
    axes[i].plot(current[1:], -1*deriv/power[1:]*100/0.04, color='k')  # -1 because current is ramped from high to low. 0.04 mA is the stepsize
    axes[i].set_ylim(-5, 5)
    axes[i].set_title(laser['name'])
    axes[i].set_xlabel('Current (mA)')
    axes[i].set_ylabel('(dP/P)/dI (%/mA)')
fig.tight_layout()
```
