```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
DATA_DIR = "data/2023-05-21.csv"
data = pd.read_csv(DATA_DIR, index_col="time")
```

```python
fig, ax = plt.subplots(dpi=300)
ax.plot(data[["I_4", "I_5", "I_6"]])
```

```python
fig, ax = plt.subplots(3, 1, dpi=200, figsize=(4, 6))

ax[0].plot(pd.to_datetime(data.index), data[["P_4"]]*1e3, '.-')
ax[0].set(
    ylim=(7.5, 7.73), 
    ylabel="2DMOT [mW]")
ax[0].tick_params(axis='x', rotation=45)

ax[1].plot(pd.to_datetime(data.index), data[["P_5"]]*1e3, '.-')
ax[1].set(
    ylim=(9.5, 9.75), 
    ylabel="MOT [mW]")
ax[1].tick_params(axis='x', rotation=45)

ax[2].plot(pd.to_datetime(data.index), data[["P_6"]]*1e3, '.-')
ax[2].set(
    ylim=(11.5, 11.73), 
    ylabel="ZS [mW]")
ax[2].tick_params(axis='x', rotation=45)

fig.tight_layout()
```
